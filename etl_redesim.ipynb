```python
import pandas as pd
import os
import warnings
from unidecode import unidecode
from layout.capitais import capitais

warnings.filterwarnings('ignore')
```

```python
archives = os.listdir('databases/xlsx')
```

```python
def load_data_excel(name_archive):

    new_name_archive = name_archive[0:-5]
    path_origin = 'databases/xlsx/'+name_archive
    path_destination = 'databases/csv/'+new_name_archive+'.csv'

    print(f'Lendo o arquivo {name_archive}')

    df = pd.read_excel(path_origin, skiprows=1)

    print(f'Exportando o arquivo {new_name_archive}.csv')
    df.to_csv(path_destination,index=False,sep=';')
    linhas_processadas = df.shape[0]
    print(f'Foram processadas {linhas_processadas} linhas')
    print('-----------------------')
```

```python
%%time
count = 0
for i in archives:
    load_data_excel(i)
    count+=1
    print(f'{count}/{len(archives)}')
```

# Processando os CSVs

```python
archives = os.listdir('databases/csv')
```

```python
def load_data(path):
    df = pd.read_csv(path, sep=';', encoding='utf-8', decimal=',')
    df = df[df.columns.drop(list(df.filter(regex='CNAE SECUNDARIA')))]
    df = df[df.columns.drop(list(df.filter(regex='FORMA ATUACAO')))]
    df['DATA DEFERIMENTO'] = pd.to_datetime(df['DATA DEFERIMENTO'],format='%d/%m/%Y', errors='ignore')
    return df
```

```python
result = []
count = 0

for i in archives:
    df = load_data('databases/csv/'+i);
    result.append(df);
    count+=1;
    print(f'{count}/{len(archives)} > Processou o arquivo {i}');
```

```python
# df = pd.concat([load_data('databases/csv/'+i) for i in archives])
```

```python
df = pd.concat(result)
```

```python
def first_word(string):
    text = str(string)
    text = unidecode(text)
    text = text.replace('.','_').replace(' ','_')
    return text.lower()
```

```python
first_word('HORA.TRANSMISSÃO')
```

```python
df.columns = [first_word(i) for i in df.columns]
```
